## Maximum connected / v1

Given a `grid 4x3`, find the maximum connected colors / use Recursion

```python
grid = [
    [0, 0, 2, 2],
    [0, 1, 2, 1],
    [2, 1, 1, 1],
]

UP, DOWN, LEFT, RIGHT = 'up', 'down', 'left', 'right'

def max_connected(A, start=(0, 0), parent=None, color=0):

    y, x = start
    m = 0

    # Base cases
    if y < 0 or x < 0: 
        return 0

    if y == len(A)-1 or x == len(A[0])-1:
        return 0

    # Neighbours coordinates
    top = y-1, x
    right = y, x+1
    bottom = y+1, x
    left = y, x-1

    # Count north
    y, x = top
    if A[y][x] == color and parent != UP: 
        m = m + 1 + max_connected(A, top, parent=DOWN)

    # Count east
    y, x = right
    if A[y][x] == color and parent != RIGHT: 
        m = m + 1 + max_connected(A, right, parent=LEFT)

    # Count south
    y, x = bottom
    if A[y][x] == color and parent != DOWN: 
        m = m + 1 + max_connected(A, bottom, parent=UP)

    # Count west
    y, x = left
    if A[y][x] == color and parent != LEFT: 
        m = m + 1 + max_connected(A, left, parent=RIGHT)

    return m

A = grid

totals = []
for j in range(len(A)):
    for i in range(len(A[0])):

        start = (j, i)
        color = A[j][i]

        m_ = max_connected(A, start, color=color)
        totals.append(m_)

print(totals)
max_connected = max(totals) + 1 # 1 for the start item itself
print(max_connected)
```

## Maximum connected / v2

Given a `grid 4x3`, find the maximum connected colors / use Iteration

```python
# We can count only DOWN and RIGHT for each element

A = [
    [0, 0, 2, 2, 2],
    [0, 0, 2, 1, 2],
    [2, 1, 1, 1, 2],
]

connections = {'0': 0, '1': 0, '2': 0}

for y in range(len(A)):
    for x in range(len(A[0])):
        color = A[y][x]

        # Right
        if x+1 < len(A[0]): 
            if A[y][x+1] == color:
                connections[str(color)] += 1

        # Down
        if y+1 < len(A): 
            if A[y+1][x] == color:
                connections[str(color)] += 1

print(connections)
max_connections = max(connections.values()) + 1 # 1 for start item itself
print(max_connections)
```

## Image Fill

Create an image with `black and white` cell and fill some connected cell / use Tkinter and PIL

### Image

Create a `blank image` using Tkinter.

```python
import tkinter as tk
from PIL import Image, ImageTk

# Create a Tkinter window
window = tk.Tk()
window.title("Image Fill")

# Create a new blank image
image = Image.new("RGB", (400, 400), "white")

# Create a Tkinter compatible photo image
photo = ImageTk.PhotoImage(image)

# Create a label widget to display the image
image_label = tk.Label(window, image=photo)
image_label.pack()

# Start the Tkinter event loop
window.mainloop()
```

### Cells

Add `black and white` cells to the image.

```python
import tkinter as tk
from PIL import Image, ImageTk

# Create a Tkinter window
window = tk.Tk()
window.title("Image Fill")

# Define dimensions
image_width = 180
image_height = 180

# Create a new blank image
image = Image.new("RGB", (image_width, image_height), "white")

# -----------------------------------------------

# Calculate the number of cells in each dimension
cell_size = 20
n_x = image_width // cell_size
n_y = image_height // cell_size

BLACK = 'green'
WHITE = 'white'

# Iterate over the cells and set their colors
for y in range(n_y):
    for x in range(n_x):

        # Calculate cell boundaries
        left = x * cell_size
        top = y * cell_size
        right = left + cell_size
        bottom = top + cell_size

        # Set the cell color based on cell position
        # color = 'black' if (x + y) % 2 == 0 else 'white' 

        # Set the color based on a model
        color = WHITE

        # Set borders
        if x == 0 or y == 0: color = BLACK
        if x == n_x -1 or y == n_y - 1: color = BLACK

        # Set middle
        if x == 4 or y == 4: color = BLACK
        if x == 4 and y > 2 and y < 6: color = WHITE
        if y == 4 and x > 2 and x < 6: color = WHITE
        if (x, y) == (3,3): color = BLACK
        if (x, y) == (5,5): color = BLACK

        # Fill the cell with color
        image.paste(color, (left, top, right, bottom))

# -----------------------------------------------

# Create a Tkinter compatible photo image
photo = ImageTk.PhotoImage(image)

# Create a label widget to display the image
image_label = tk.Label(window, image=photo)
image_label.pack()

# Start the Tkinter event loop
window.mainloop()
```
